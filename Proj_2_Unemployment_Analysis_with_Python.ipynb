# Unemployment Analysis with Python - Pakistan

**Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force**

The unemployment rate is calculated based on a particular region, so to analyze unemployment I will be using an unemployment dataset of Pakistan.

```python
# Importing Neccesary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Reading Data
data = pd.read_csv("/content/Pakistan.csv")
data.head()
```

```python
data.tail()
```

```python
print(data.isnull().sum())
```

```python
# Correlation between different features

plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
data.info()
```

# Let's do some EDA

**1. Time Series / Trend Analysis**

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='GDP Growth Rate (%)', data=data)
plt.title('GDP Growth Rate Over Time')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Inflation Rate (%)', data=data)
plt.title('Inflation Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Poverty Headcount Ratio (%)', data=data)
plt.title('Poverty Headcount Ratio Over Time')
plt.xlabel('Year')
plt.ylabel('Poverty Headcount Ratio (%)')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Time Series Analysis of Various Factors', fontsize=16)

sns.lineplot(x='Year', y='Agriculture Growth Rate (%)', data=data, ax=axes[0, 0])
axes[0, 0].set_title('Agriculture Growth Rate Over Time')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Agriculture Growth Rate (%)')

sns.lineplot(x='Year', y='Government Social Spending (% of GDP)', data=data, ax=axes[0, 1])
axes[0, 1].set_title('Government Social Spending Over Time')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Government Social Spending (% of GDP)')

sns.lineplot(x='Year', y='External Debt (USD billions)', data=data, ax=axes[1, 0])
axes[1, 0].set_title('External Debt Over Time')
axes[1, 0].set_xlabel('Year')
axes[1, 0].set_ylabel('External Debt (USD billions)')

sns.lineplot(x='Year', y='Climate Disasters (count)', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Climate Disasters Over Time')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Climate Disasters (count)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**2. Correlation & Relationship Analysis**

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Relationship Analysis', fontsize=16)

sns.scatterplot(x='Poverty Headcount Ratio (%)', y='Unemployment Rate (%)', data=data, ax=axes[0])
axes[0].set_title('Unemployment Rate vs Poverty Headcount')
axes[0].set_xlabel('Poverty Headcount Ratio (%)')
axes[0].set_ylabel('Unemployment Rate (%)')

sns.scatterplot(x='GDP Growth Rate (%)', y='Inflation Rate (%)', data=data, ax=axes[1])
axes[1].set_title('Inflation Rate vs GDP Growth Rate')
axes[1].set_xlabel('GDP Growth Rate (%)')
axes[1].set_ylabel('Inflation Rate (%)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP Growth Rate (%)', y='External Debt (USD billions)', data=data)
plt.title('External Debt vs GDP Growth Rate')
plt.xlabel('GDP Growth Rate (%)')
plt.ylabel('External Debt (USD billions)')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Distribution of Key Variables', fontsize=16)

sns.histplot(data['Unemployment Rate (%)'], kde=True, ax=axes[0])
axes[0].set_title('Distribution of Unemployment Rate (%)')
axes[0].set_xlabel('Unemployment Rate (%)')
axes[0].set_ylabel('Frequency')

sns.histplot(data['Inflation Rate (%)'], kde=True, ax=axes[1])
axes[1].set_title('Distribution of Inflation Rate (%)')
axes[1].set_xlabel('Inflation Rate (%)')
axes[1].set_ylabel('Frequency')

sns.histplot(data['External Debt (USD billions)'], kde=True, ax=axes[2])
axes[2].set_title('Distribution of External Debt (USD billions)')
axes[2].set_xlabel('External Debt (USD billions)')
axes[2].set_ylabel('Frequency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Box Plots to Detect Outliers', fontsize=16)

sns.boxplot(y='Inflation Rate (%)', data=data, ax=axes[0])
axes[0].set_title('Inflation Rate (%)')
axes[0].set_ylabel('Inflation Rate (%)')

sns.boxplot(y='External Debt (USD billions)', data=data, ax=axes[1])
axes[1].set_title('External Debt (USD billions)')
axes[1].set_ylabel('External Debt (USD billions)')

sns.boxplot(y='Climate Disasters (count)', data=data, ax=axes[2])
axes[2].set_title('Climate Disasters (count)')
axes[2].set_ylabel('Climate Disasters (count)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**So this is how you can analyze the unemployment rate by using the Python programming language. Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force.**
